```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
```

```python
url = 'https://raw.githubusercontent.com/asukul/DS301-f19/master/datasets/credit_data/german_credit_data.csv'
df = pd.read_csv(url)
```

```python
df = df.drop(columns=['Saving accounts', 'Checking account', 'Unnamed: 0'])

df.fillna(df.mean(), inplace=True)
```

```python
df = df.fillna(df.mode().iloc[0])
df = pd.concat([df, pd.get_dummies(df['Sex'], prefix = 'Sex'),
                pd.get_dummies(df['Job'], prefix = 'Job'),
                pd.get_dummies(df['Housing'], prefix = 'Housing'),
                pd.get_dummies(df['Purpose'], prefix = 'Purpose')],axis=1)
df = df.drop(columns=['Sex', 'Job', 'Housing', 'Purpose'])
df.head()
```

```python
le = LabelEncoder()

df['Risk'] = le.fit_transform(df['Risk'])
df.head()
```

```python
y = df.Risk.values
x = df.drop(columns=['Risk'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
# Decision Tree
clf = DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

decisionTreeScore = accuracy_score(y_test, y_pred)

# Naive Bayes
clf = GaussianNB()
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

naiveBayesScore = accuracy_score(y_test, y_pred)

# KNN
clf = KNeighborsClassifier(n_neighbors=5)
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

knnScore = accuracy_score(y_test, y_pred)

# LogReg
clf = LogisticRegression()
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

logRegScore = accuracy_score(y_test, y_pred)

# SVC
clf = SVC()
clf = clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

svcScore = accuracy_score(y_test, y_pred)
```

```python
# Standard Scaler
scaler = StandardScaler()
scaler.fit(x_train)

x_trainScaler = scaler.transform(x_train)
x_testScaler = scaler.transform(x_test)
```

```python
# Decision Tree
clf = DecisionTreeClassifier()
clf = clf.fit(x_trainScaler, y_train)

y_pred = clf.predict(x_testScaler)

decisionTreeScalerScore = accuracy_score(y_test, y_pred)

# Naive Bayes
clf = GaussianNB()
clf = clf.fit(x_trainScaler, y_train)

y_pred = clf.predict(x_testScaler)

naiveBayesScalerScore = accuracy_score(y_test, y_pred)

# KNN
clf = KNeighborsClassifier(n_neighbors=5)
clf = clf.fit(x_trainScaler, y_train)

y_pred = clf.predict(x_testScaler)

knnScalerScore = accuracy_score(y_test, y_pred)

# LogReg
clf = LogisticRegression()
clf = clf.fit(x_trainScaler, y_train)

y_pred = clf.predict(x_testScaler)

logRegScalerScore = accuracy_score(y_test, y_pred)

# SVC
clf = SVC()
clf = clf.fit(x_trainScaler, y_train)

y_pred = clf.predict(x_testScaler)

svcScalerScore = accuracy_score(y_test, y_pred)
```

```python
print('Non Scaler Score')
print('Decision Tree Score: ' + str(decisionTreeScore))
print('Naive Bayes Score: ' + str(naiveBayesScore))
print('KNN Score: ' + str(knnScore))
print('LogReg Score: ' + str(logRegScore))
print('SVC Score: ' + str(svcScore))
print('')
print('Scaler Score')
print('Decision Tree Score: ' + str(decisionTreeScalerScore))
print('Naive Bayes Score: ' + str(naiveBayesScalerScore))
print('KNN Score: ' + str(knnScalerScore))
print('LogReg Score: ' + str(logRegScalerScore))
print('SVC Score: ' + str(svcScalerScore))
```

```python
# For some classifiers, the scaler score seems to perform better
```

```python
# Random Forest
clf = RandomForestClassifier(n_estimators=1000)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

randomForestScore = accuracy_score(y_test, y_pred)

# AdaBoost
clf = AdaBoostClassifier()
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

adaBoostScore = accuracy_score(y_test, y_pred)

# XGBoost
clf = XGBClassifier()
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

xgbScore = accuracy_score(y_test, y_pred)
```

```python
print('Non Scaler Score')
print('Random Forest Score: ' + str(randomForestScore))
print('Adaboost Score: ' + str(adaBoostScore))
print('XGBoost Score: ' + str(xgbScore))
```

```python
clf = VotingClassifier(estimators=[('dt', DecisionTreeClassifier()), ('nb', GaussianNB()), ('knn', KNeighborsClassifier(n_neighbors=5)),
                                   ('lr', LogisticRegression()), ('svc', SVC()), ('rf', RandomForestClassifier(n_estimators=1000)),
                                   ('ab', AdaBoostClassifier()), ('xgb', XGBClassifier())], voting='hard')
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)

votingScore = accuracy_score(y_test, y_pred)
```

```python
print('Voting Score: ' + str(votingScore))
```
