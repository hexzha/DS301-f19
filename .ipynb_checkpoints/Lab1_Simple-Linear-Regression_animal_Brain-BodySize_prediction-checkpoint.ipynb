```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("datasets/Brain-BodySize/brain-bodysize.csv")
dataset.describe()
```

```python
x = dataset[['Brain']]
y = dataset[['Body']]

dataset
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#using scikit-learn to train your model with LinearRegression.
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
print(model.intercept_)
print(model.coef_)

y_pred = model.predict(x_test)
```

```python
#use matplotlib (or any visualization libraries) to make the plot describing this data.
#add your regression line to your plot, if you haven't already done so

dataset.plot(x='Brain', y='Body', style='o')  
plt.title('Brain vs Body')  
plt.xlabel('Brain')  
plt.ylabel('Body')  
plt.plot(x_test, y_pred, color='red', linewidth=2)
plt.show()
```

```python
#create three different new unseen data points (Xs) and make predictions from your model (Ys)
x_new = [[300]]
print("For 300: " + str(model.predict(x_new)[0][0])) 

x_new = [[400]]
print("For 400: " + str(model.predict(x_new)[0][0])) 

x_new = [[450]]
print("For 450: " + str(model.predict(x_new)[0][0])) 
```
